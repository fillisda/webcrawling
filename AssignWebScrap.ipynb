**Question 2: Use Webscraping to Extract Tesla Revenue Data**

***Display the last five rows of the tesla_revenue dataframe using the tail function. Upload a screenshot of the results.***

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://stockanalysis.com/stocks/tsla/revenue/'
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
import pandas as pd
from bs4 import BeautifulSoup

# Assuming soup is already defined
# Initialize an empty DataFrame
revenue_data = pd.DataFrame(columns=["Fiscal Year End", "Revenue", "Change", "Growth"])

for row in soup.find("tbody").find_all('tr'):
    columns = row.find_all("td")
    fiscal_year_end = columns[0].text.strip()
    revenue = columns[1].text.strip()
    change = columns[2].text.strip()
    growth = columns[3].text.strip()
    
    # Append the data of each row to the table
    revenue_data = pd.concat(
        [revenue_data, 
         pd.DataFrame({"Fiscal Year End": [fiscal_year_end], "Revenue": [revenue], "Change": [change], "Growth": [growth]})
        ], ignore_index=True)

print(revenue_data)
```
